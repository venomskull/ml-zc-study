```python
X = data[data['origin'] == 'Asia'][['vehicle_weight', 'model_year']][:7]
X
```

```python
XTX = X.T.dot(X)
XTX
```

```python
XTX_inv = np.linalg.inv(XTX)
XTX_inv
```

```python
y = np.array([1100, 1300, 800, 900, 1000, 1100, 1200])
y
```

```python
w = (XTX_inv @ X.T) @ y
w
```
